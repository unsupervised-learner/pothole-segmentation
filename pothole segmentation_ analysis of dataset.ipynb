```python
import PIL.Image as img
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

## testing the conversion of mask to binary mask for random image from training set

```python
#define paths to image directories
main_path = 'C:/Users/android18/Pictures/potholes'
image_path = os.path.join(main_path, 'images')
mask_path = os.path.join(main_path, 'masks')
binary_mask_path = os.path.join(main_path, 'binary_masks')
```

```python
#load random image from training data
rand_ind = np.random.randint(3)
targ = str(rand_ind)+'.png'
image = cv.imread(os.path.join(image_path, targ), 0)
binary_mask = cv.imread(os.path.join(binary_mask_path, targ), 0)
mask = binary_mask < 215
mask2 = binary_mask >= 215
binary_mask[mask] = 0
binary_mask[mask2] = 1


fig = plt.figure(facecolor='pink', figsize=(10,20))

fig.add_subplot(1,2,1)
plt.imshow(image)
plt.axis('off')

fig.add_subplot(1,2,2)
plt.imshow(binary_mask)
plt.axis('off')


plt.show()
```

## Convert images and masks to standard size

```python
#convert and save all masks
for image in os.listdir(binary_mask_path):
    im = cv.imread(os.path.join(binary_mask_path, image), 0)
    mask1 = im < 215
    mask2 = im >= 215
    im[mask1] = 0
    im[mask2] = 1
    im = cv.resize(im, (600,600), cv.INTER_AREA)
    plt.imsave(os.path.join(binary_mask_path, image), im)
```
